```python
def build_and_compile_model():
  model = keras.Sequential([
#       norm,
      layers.InputLayer(shape=(8,)),
      layers.Dense(32, activation='relu'),
      layers.Dense(16, activation='relu'),
      layers.Dense(1)
  ])

  model.compile(loss='mean_absolute_error',
                optimizer=tf.keras.optimizers.Adam(learning_rate=0.0005,
                                                    beta_1=0.9,
                                                    beta_2=0.999,
                                                    epsilon=1e-07,
                                                    amsgrad=False,
                                                    weight_decay=None,
                                                    clipnorm=None,
                                                    clipvalue=None,
                                                    global_clipnorm=None,
                                                    use_ema=False,
                                                    ema_momentum=0.99,
                                                    ema_overwrite_frequency=None,
                                                    loss_scale_factor=None,
                                                    gradient_accumulation_steps=None,
                                                    name='adam'
                                                ))
  return model
```

```python
scaler_input = MinMaxScaler()
scaled_train_features=scaler_input.fit_transform(train_features)

scaler_output = MinMaxScaler()
scaled_train_labels=scaler_output.fit_transform(train_labels.reshape(-1,1))

dnn_model = build_and_compile_model()
dnn_model.summary()
```

```python
%%time
history = dnn_model.fit(
    scaled_train_features,
    scaled_train_labels,
    validation_split=0.2,
    verbose=0, epochs=250)
```

```python
def plot_loss(history):
  plt.plot(history.history['loss'], label='loss')
  plt.plot(history.history['val_loss'], label='val_loss')
  plt.ylim([0.1, 0.2])
  plt.xlabel('Epoch')
  plt.ylabel('Error [Cost]')
  plt.legend()
  plt.grid(True)
```

```python
plot_loss(history)
```

```python
# test_results
scaled_test_features=scaler_input.fit_transform(test_features)
scaled_test_labels=scaler_output.fit_transform(test_labels.reshape(-1,1))


test_results = dnn_model.evaluate(scaled_test_features, scaled_test_labels, verbose=0)
```
